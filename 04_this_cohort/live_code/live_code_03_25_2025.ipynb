```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

Null hypothesis --> no significant linear relationship between house size and house price (b1 = 0)

Alternative hypothesis --> there is a significant linear relationship between house size and house price (b1 != 0)

```python
sacramento = pd.read_csv("dataset/sacramento.csv")
sacramento
```

```python
# Plot the relationship between x and y
plt.scatter(sacramento["sq__ft"], sacramento["price"], label='House size vs House price')
```

```python
# model(y ~ x, data).fit()
```

```python
#simple linear regression

model = smf.ols("price ~ sq__ft", data = sacramento).fit()
print(model.summary())
```

```python
# Residual plots
fitted = model.fittedvalues
residuals = model.resid


# Homoscedasticity: Residuals vs. Fitted
plt.scatter(fitted, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted')
plt.show()
```

```python
sm.qqplot(residuals, line = 'q')
plt.title('Q-Q Plot')
plt.show()
```

```python
#multivariable linear regression
multi_model = smf.ols("price~ sq__ft + beds", data = sacramento).fit()
print(multi_model.summary())
```
